```python
np.random.seed(0)

creator.create('FitnessMin', base.Fitness, weights=(-1.0,))
creator.create('Individual', list, fitness=creator.FitnessMin)
```

```python
TOURN_SIZE = 10 #Size of tournament block

#Initialization of our random selector
toolbox = base.Toolbox()


#Operators
#Crossover
toolbox.register('mate', tools.cxTwoPoint)
#Mutation
toolbox.register('mutate', tools.mutShuffleIndexes, indpb=0.05)
#Selection
toolbox.register('select', tools.selTournament, tournsize=TOURN_SIZE)
```

```python
def GA(POP_SIZE, CXPB, MUTPB, NGEN, stats, start, end, IND_SIZE=IND_SIZE, starting_point=starting_point, ending_point=ending_point): #other than the usual arguments, the function takes also the weights and values specified and the size of our problem

    
    size = IND_SIZE
    active_list = ALL_LIST.copy()
    
    if starting_point == "":
        starting_point = start
    else:
        starting_point = starting_point
    #active_list.append(starting_point)
    size +=1
    
    if ending_point == "":
        ending_point = end
        #active_list.append(ending_point)
        size +=1
    else:
        ending_point = ending_point
    #print(size)
        
    #print(active_list)

    
        
    #Defininf Hall of Fame
    hof = tools.HallOfFame(1)
    def attr_string():
        r = random.sample(active_list, size - 2)
        r.append(starting_point)
        r.append(ending_point)
        return r
    #print(attr_string())
    
    #HOF = [] #List created in order to store the best-weight (the maximum under the treshold) processed for every generation
    toolbox.register('attr_string', attr_string)
    

    #Creating the population
    #individuals
    toolbox.register('individual', tools.initIterate, creator.Individual, toolbox.attr_string) #here we can decide which size of our problem we are considering
    #population
    toolbox.register('population', tools.initRepeat, list, toolbox.individual)
    
    pop = toolbox.population(n=POP_SIZE)
    #print(pop)
    for i in pop:
        counter = 0
        for j in i:
            if j ==("Entrance 1", "Structure"):
                counter+=1
        if counter>=2:
            print(i)
            
    
    #print([starting_point for i in range(len(pop))])
    toolbox.register('evaluate', evaluate)

    #Defining the Logbook
    logbook = tools.Logbook()
    logbook.header = ["gen", "nevals"] + (stats.fields if stats else[])

    #Evaluate the entire population
    fitness = list(map(toolbox.evaluate, pop, [starting_point for i in range(len(pop))], [ending_point for i in range(len(pop))]))
    for ind, fit in zip(pop, fitness):
        ind.fitness.values = [fit]



    hof.update(pop) if stats else {}

    record = stats.compile(pop) if stats else {}
    logbook.record(gen=0, nevals=len(pop), **record)
    

    for g in range(NGEN):
        #Select the next generation individuals
        offspring = toolbox.select(pop, len(pop))
        #Clone the selected individuals
        offspring = list(map(toolbox.clone, offspring))

        #Apply crossover and mutation on the offspring
        for child1, child2 in zip(offspring[::2],offspring[1::2]):
            if random.random() < CXPB:
                child1.remove(starting_point)
                child1.remove(ending_point)
                child2.remove(starting_point)
                child2.remove(ending_point)
                toolbox.mate(child1,child2)
                del child1.fitness.values
                del child2.fitness.values
                toolbox.mate(child1,child2)[0].insert(0,ending_point)
                toolbox.mate(child1,child2)[0].append(starting_point)
                toolbox.mate(child1,child2)[1].insert(0,ending_point)
                toolbox.mate(child1,child2)[1].append(starting_point)
                
        for mutant in offspring:
            if random.random() < MUTPB:
                toolbox.mutate(mutant)
                del mutant.fitness.values

        #Evaluate the individuals with an invalid fitness
        invalid_ind = [ind for ind in offspring if not ind.fitness.valid]
        for i in invalid_ind:
            counter = 0
            for j in i:
                if j ==("Entrance 1", "Structure"):
                    counter+=1
            if counter>=2:
                print(i)
        fitness = list(map(toolbox.evaluate, invalid_ind, [starting_point for i in range(len(pop))], [ending_point for i in range(len(pop))]))   
        for ind, fit in zip(invalid_ind, fitness):
            ind.fitness.values = [fit]
        if hof is not None:
            hof.update(offspring)

        #The population in entirely replaced by the offspring
        pop[:] = tools.selBest(offspring, POP_SIZE-1)
        pop.append(hof[0])
        #HOF.append(hof[0]) #here, for instance, we add the best individual for eache iteration (generation) to the HOF list created above


        record = stats.compile(pop) if stats else{}
        logbook.record(gen=g+1, nevals=len(invalid_ind), **record)
        
        
    fitness_best = list(map(toolbox.evaluate, pop, [starting_point for i in range(len(pop))], [ending_point for i in range(len(pop))]))
    for fit in fitness_best:
        if fit == min(fitness_best):
            index = fitness_best.index(fit)
    best = pop[index][0]
        
    return pop, logbook, best

```
